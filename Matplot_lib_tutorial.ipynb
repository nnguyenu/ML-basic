```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
x = np.arange(0, 100, 0.1)
y = np.sin(x)
plt.plot(x, y)
```

General format for plot:

plot([x], y, [fmt], *, data=None, **kwargs) 

The optional parameter fmt is a convenient way for defining basic formatting like color, marker and linestyle. It's a shortcut string notation described in the Notes section below.

```python
plt.plot(x, y, 'bo')  # plot x and y using blue circle markers
```

```python
N = 6
wave = [940,880,605,507,505,470]
v = [1.19,1.29,1.89,2.66,2.59,2.73]
c = 299792458
e = 1.60217662 * 10**(-19)
h =  6.626070040 * (10**(-34))
p = [0]*N

print('655 nm:',e*1.648*655*(10**(-9))/c)

for i in range(N): 
  p[i] = e*v[i]*wave[i]*(10**(-9))/c
print(p)

ex = [0]*N
for i in range(N):
  ex[i] = c*h/wave[i]/e*(10**9)

wrong = [0]*N
for i in range(N):
  wrong[i] = round(v[i]/ex[i]*100,2)

mean = sum(p)/len(p)
print('mean:',mean)

std = 0
for i in range(N):
  std += (p[i]-mean)**2/N 
std = math.sqrt(std)
print('std:',std)
uncertain = std/mean
print('uncertain:',uncertain)

relative = (mean-h)/h
print('relative:',relative)
rede = [0]*N
for i in range(N):
  rede[i] = (p[i]-h)/h
print('rede:',rede) 

plt.plot(wave,p,'ro')

x = [450,950]
y = [h]*2
yy = [mean]*2
plt.plot(x,y)
plt.plot(x,yy,'green')

yd1 = [mean-std]*2
yd2 = [mean+std]*2

plt.fill_between(x,yd1,y,color = 'yellow')
plt.fill_between(x,yd2,y,color = 'yellow')

plt.rcParams.update({'font.size': 9.7})
plt.xlabel('Wavelength (nm)')
plt.ylabel("Planck's constant estimation (J.s)")
plt.title("Planck's constant as a function of wavelength")
```
